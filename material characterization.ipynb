```python
# create a class  for frequency encoder
class FrequencyEncoder:
    def __init__(self):
        self.frequency_map = {}

    def fit(self, X, y=None):
        for column in X.columns:
            self.frequency_map[column] = X[column].value_counts(normalize=True).to_dict()
        return self

    def transform(self, X):
        X_encoded = X.copy()
        for column in X.columns:
            X_encoded[column] = X[column].map(self.frequency_map[column])
        return X_encoded   
```

```python
from sklearn.preprocessing import LabelEncoder


# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate over each column and encode if it's categorical
for column in y_train.columns:
    if y_train[column].dtype == 'object':  # Check if column is categorical
        y_train[column] = label_encoder.fit_transform(y_train[column])
```

```python
# Create an instance of FrequencyEncoder
encoder = FrequencyEncoder()

# Fit the encoder to the training data
encoder.fit(df[categorical_features])

# Transform the training and test data using the encoder
x_train_encoded=encoder.transform(x_train[x_train_categorical_features])

x_test_encoded=encoder.transform(x_test[x_train_categorical_features])
```

```python
# now handle the numerical features
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train[numerical_features] = scaler.fit_transform(x_train[numerical_features])
x_test[numerical_features] = scaler.transform(x_test[numerical_features])
```

```python
# Concatenate numerical and categorical features in x_train and x_test
x_train_concat = pd.concat([x_train[numerical_features], x_train_encoded], axis=1)
x_test_concat = pd.concat([x_test[numerical_features], x_test_encoded], axis=1)
```

```python
# Ensure all features are of the correct data type
x_train_concat = x_train_concat.astype(float)
x_test_concat = x_test_concat.astype(float)
```
